# Exploratory Data Analysis on Dataset - Terrorism

The aim of this Project is to find out the hot zone of terrorism from a data of Terrorism Occured Every Year. The Data is given in .csv file.

Dataset: https://bit.ly/2TK5Xn5

Presenting an Analysis of the data by
1. Year Wise Attacks
2. Year Wise Regional Attacks
3. Year Wise Attack Type Attacks
4. Year Wise Target Type Attacks
5. Year Wise Weapon Type Attacks
6. Year Wise Country Attacks
7. Motive of The Attacks



```python
#ALL THE LIBRARIES REQUIRED
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
#DATA COLLECTION
data=pd.read_csv('globalterrorismdb_0718dist.csv',encoding='latin1')
df=pd.DataFrame(data)
```

```python
#ALL THE REQUIRED COLUMNS FOR ANALYSIS
df=df[['iyear' , 'imonth' , 'iday' ,'country_txt','region_txt', 'provstate' , 'city' ,'latitude','longitude','location','summary','attacktype1_txt','targtype1_txt','gname','motive','weaptype1_txt','nkill','nwound','addnotes']]
df.head()
```

```python
df.value_counts('city')
```

```python
#To Check NuLL Entries
df.isnull().sum()
```

```python
#HANDLING MISSING VALUES
df['nkill']=df['nkill'].fillna(0)
df['nwound']=df['nwound'].fillna(0)
df['Affected']=df['nkill']+df['nwound']
```

# 1)Year wise Attacks

```python
a1= df['iyear'].value_counts(dropna=False)
a2= a1.sort_index()
a3= a2.to_frame().reset_index().rename(columns={'index':'iyear','iyear':'Attacks'})
attacks= a3.set_index('iyear')
print(attacks)
```

```python
sns.jointplot(x='iyear',y='Attacks',data=attacks,kind='scatter')
```

```python
attacks.plot(kind='bar',figsize=(13,6),fontsize=10)
plt.title('Attacks Every Year',fontsize=13)
plt.xlabel('Years',fontsize=13)
plt.ylabel('Attacks',fontsize=13)
plt.show()
```

*The most of the attacks happened to be in 2014(Which is 16903).

*The least number of attacks happened to be in 1971(Which is 471).

```python
#YEAR WISE AFFECTED
YC=df[['iyear','Affected']].groupby('iyear').sum()
YC
```

```python
YC.plot(kind='bar',figsize=(12,6))
plt.title('People Got both Killed and Wounded (Affected) every Year')
plt.xlabel('Year',fontsize=12)
plt.ylabel('Affected',fontsize=12)
plt.show()

```

*Most of the people got affected in 2014(Which is 85618)

*Least of the people got affected in 1971(which is 255)

```python
#YEAR WISE GOT KILLED
YK=df[['iyear','nkill']].groupby('iyear').sum()
YK
```

```python
YK.plot(kind='bar',figsize=(12,6))
plt.title('People Got Killed every Year')
plt.xlabel('Year',fontsize=12)
plt.ylabel('Killed',fontsize=12)
plt.show()
```

*Most of the people got Killed in 2014(Which is 44490)

*Least of the people got Killed in 1971(which is 173))

```python
#YEAR WISE GOT WOUNDED
YW=df[['iyear','nwound']].groupby('iyear').sum()
YW
```

```python
YW.plot(kind='bar',figsize=(12,6))
plt.title('People Got Wound every Year')
plt.xlabel('Year',fontsize=12)
plt.ylabel('Wounded',fontsize=12)
plt.show()
```

*Most of the people got wounded in 2015(Which is 44043)

*Least of the people got wounded in 1971(which is 82)

# 2) Year Wise Regional Attacks

```python
YR=pd.crosstab(df.iyear,df.region_txt)
YR.head()
```

```python
YR.plot(kind='area',figsize=(12,6),)
plt.title('Attacks in Every Region in between 1970-2017')
plt.xlabel('Year')
plt.ylabel('Attacks')
plt.show()
```

```python
YRT=YR.transpose()
YRT['Total']=YRT.sum(axis=1)
activities=YRT['Total'].sort_values(ascending=False)
activities
```

```python
activities.dtypes
```

```python
activities.plot(kind='bar',figsize=(12,6))
plt.title('number of attacks in each region in between 1970-2017')
plt.xlabel('Region')
plt.ylabel('Attacks')
plt.show()
```

*Most of the Attacks where happened in Middle East & North Africa(i.e 50474)

*Least of the Attacks where happened in Australasia & Oceania(i.e 282)

```python
#REGION WISE AFFECTED
RC=df[['region_txt','Affected']].groupby('region_txt').sum()
RC
```

```python
RC.plot(kind='bar',figsize=(12,6))
plt.title('Affected in each region',fontsize=10)
plt.xlabel('Region',fontsize=10)
plt.ylabel('Affected',fontsize=10)
plt.show()
```

*Most of the people got Affected in Middle East & North Africa(i.e 351950)

*Least of the people got Affected in Australasia & Oceania(i.e 410)

```python
#REGION WISE KILLED
RK=df[['region_txt','nkill']].groupby('region_txt').sum()
RK
```

```python
RK.plot(kind='bar',figsize=(12,6))
plt.title('No. of People Killed in each region',fontsize=10)
plt.xlabel('Region',fontsize=10)
plt.ylabel('Killed',fontsize=10)
plt.show()
```

*Most of the people got Killed in Middle East & North Africa(i.e 137642)
