# Iris Flowers Classification ML Project

The Motive of this Project is to design Model which classify iris flowers in three types('Iris Setosa','Iris Versicolour',
'Iris Virginica')through the given measurements Sepas and Petals(length and Width).


Data Source used for this classification project has '5' Columns and '150' Rows.It Contains Data in .csv file.

Attribute Information of Iris Flower Dataset:

1.Sepal Length in cm   
2.Sepal Width in cm  
3.Petal Length in cm  
4.Petal Width in cm  
5.class: 
-Iris Setosa 
-Iris Versicolour 
-Iris Virginica 

```python
#REQUIRED LIBRARIES
import pandas as pd
import seaborn as sns
```

```python
#LOAD THE DATA
df=pd.read_csv('iris.data')
df.head()
```

```python
#ADDING HEADER TO COLUMNS
columns =['Sepal_Length','Sepal_width','Petal_Length','Petal_width', 'Class_name']
cf=pd.read_csv('iris.data',names=columns)
```

```python
#DATA VISUALIZATION
sns.pairplot(cf,hue='Class_name',palette='bright') 
```

```python
#SEPARATE FEATURE AND TARGET 
X=cf[['Sepal_Length','Sepal_width','Petal_Length','Petal_width']]
y=cf['Class_name']
```

```python
X
```

```python
y
```

```python
#TRAIN-TEST-SPLIT
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#TRAINING THE MODEL1
#MODEL1
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
#PREDICTING MODEL1
prediction=lr.predict(X_test)
```

```python
#EVALUATING MODEL1
from sklearn.metrics import accuracy_score , classification_report
print(accuracy_score(y_test,prediction)*100)
print(classification_report(y_test,prediction))
```

```python
#MODEL2
from sklearn.svm import SVC
```

```python
#TRAINING THE MODEL2
svc = SVC()
svc.fit(X_train,y_train)
```

```python
#PREDICTING MODEL2
#EVALUATING MODEL2
prediction1 = svc.predict(X_test)
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_test,prediction1)*100)
print(classification_report(y_test,prediction1))
```
